1. А точно нужно сортировать по вероятностям?

Скорей всего нет, так как некоторые товары могут быть практически идентичными, скорей всего нужно подавать группу похожих, что бы у человека была возможность выбрать среди товаров из одной категории но различающихся.

2. Какую метрику использовать?

Сложный, комплексный вопрос, надо смотреть какие метрики будут лучше работать, и лучше подходить.

3. Сколько раз в неделю рассылку?

Рассылка должна быть не навящевой, но при этом приходить максимально часто. Возможно ее можно как то встроить в рассылаемые "выжные письма" в виде полки. Так же можно перфенефецировать, в зависимости от того как давно они посещали магазин, Например если не заходили полторы недели, отправлять рассылку. А тем кто ходит часто не слать, так как они и так видят рекомендации на сайте.

4. В какое время отправляем рассылку?

Тут все сложно, но во время отдыха люди вряд ли будут сильно рады рассылку, у них обычно есть интересные дела, а вот в начале рабочего дня когда, все маятся чем бы заняться пока еще весь рабочий день в в переди, можно и почитать. Но вообще все это крайне ситуативно для каждого человека. Скорей всего нужно отправлять в разное время и смотреть в какой момент был наилучший отклик, и дальше смотреть от чего он зависит. Так как возможно разные рекомендации имеет смысл отправлять в разное время.

5. Будем отправлять одному юзеру много раз наши рекоммендации. Как добиться того, чтобы они хоть немного отличались?

Мы составляем какой-то топ рекомендаций, и скорей всего нужно тянуть из этого топа. Посылать другую часть этого топа, пока в нем не произойдут серьезные изменения за счет покупок. Плюс можно добавлять популярные сезонные товары.

6. Нужно ли, чтобы в одной рассылке были *разные* товары? Как определить, что товары *разные*? Как добиться того, чтобы они были разными?

Одинаковых быть не должно, скорей всего, форма заполнения товара должна иметь какие-то обязательные поля, которые помогут нам разделять их в нащих фичах. Вообще это не кажется простой задачей. Можно сдлеать так, что бы в одну рекомендацию не попадали товары с векторами отличающиеся меньше чем на какое то определенное эпсилон.

7. **3 товара с акцией** (Как это учесть? А если на товар была акция 10%, а потом 50%, что будет стоять в user-item матрице?)

Не уверен что это правильно относить к нашей матрице, возможно потом можно уже имея топ рекомендаций продвигать товары со скидкой.


8. **1 новый товар** (юзер никогда не покупал. Просто фильтруем аутпут ALS? А если у таких товаров очень маленькая вероятность покупки? Может, использовать другую логику/модель?) 

Можно выдавать товары рандомом из популярных, как только человек начинает посещать разделы, начинать выдавать топ популярных по ним. а дальше у нас начнеут уже строиться вектор на него.

9. **1 товар для роста среднего чека** (товары минимум дороже чем обычно покупает юзер. Как это измерить? На сколько дороже?)

Скорей всего процентно дороже, можно провести A/B тестирование по разным процентам например 10%/20%/30% на кого из них чаще тыкают. Причем учитывая, что есть цена обычно на ссылке нажатие можно считать за успех, так как человек уже согласился посомтреть, что там за эту цену лежит.
